```python
import os

import PIL.Image as Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
from stainaug.augment import Augmentor
```

```python
input_dir = '../data/'
img_fps = sorted([os.path.join(input_dir, fp)
                  for fp in os.listdir(input_dir)])
img_fps
```

```python
img = np.asarray(Image.open(img_fps[-1]))
plt.imshow(img)
```

```python
augmentor = Augmentor()
```

```python
plt.imshow(augmentor.augment_HE(img))
```

```python
plt.imshow(augmentor.augment_HE(img))
```

```python
n = 5
fig, axs = plt.subplots(ncols=len(img_fps), nrows=n, figsize=(10, 10))

for i, fp in enumerate(img_fps):
    img = np.asarray(Image.open(img_fps[i]))
    axs[0, i].imshow(img)
    axs[0, i].set_title(fp.split('/')[-1])
    
    for r in range(1, n, 1):
        axs[r, i].imshow(augmentor.augment_HE(img))
    
    for ax in axs[:, i]:
        ax.set_xticks([])
        ax.set_yticks([])
axs[0, 0].set_ylabel('Original')
```
